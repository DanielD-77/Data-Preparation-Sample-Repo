```python
import pandas as pd
import numpy as np
```

```python
mergedData = pd.read_csv("./mergedDataset.csv")
display(mergedData)
```

```python
# Drop key
mergedData = mergedData.drop("key", axis = 1)
display(mergedData)
```

# Converting fuel_type to numerical


```python
# Get uniques of fuel_type
fuel_type = mergedData["fuel_type"].unique()
print(fuel_type)
```

```python
# Convert the fuel_type column from categorical to numerical using pd.get_dummies
mergedData = pd.get_dummies(mergedData, columns = ["fuel_type"])

mergedData = mergedData.drop("fuel_type_Hybrid", axis = 1)
# When fuel_type_Diesel and fuel_type_Gasoline are both false, then it is hybrid
display(mergedData)
```

```python
mergedData['fuel_type_Diesel'] = mergedData['fuel_type_Diesel'].astype(int)
mergedData['fuel_type_Gasoline'] = mergedData['fuel_type_Gasoline'].astype(int)
display(mergedData)
```

# Converting make to numerical



```python
# Get uniques of make
make = mergedData["make"].unique()
print(make)
```

```python
# Show the frequency of each make
freqMake = mergedData['make'].value_counts()
display(freqMake)
```

```python
# Frequency encoding
frequencyDictMake = mergedData['make'].value_counts().to_dict()
mergedData['make_count'] = mergedData['make'].map(frequencyDictMake)
display(mergedData)
```

# Converting model to numerical

```python
# Get uniques of model
model = mergedData["model"].unique()
print(model)
```

```python
# Show frequency of each model
freqModel = mergedData['model'].value_counts()
display(freqModel)
```

# Converting transmission to numerical



We are still thinking whether we use frequency encoding, or use dummy variables/one-hot encoding after dropping Dual-clutch, E-cvt, Amt, and Ivt rows.

```python
# Get uniques of transmission
transmission = mergedData["transmission"].unique()
print(transmission)
```

```python
# Convert all "Cvt" into "CVT"
mergedData["transmission"] = mergedData["transmission"].replace("Cvt", "CVT")
transmission = mergedData["transmission"].unique()
print(transmission)
```

```python
# Show the frequency of each make
freqTrans = mergedData['transmission'].value_counts()
display(freqTrans)
```

```python
# Frequency encoding
frequencyDictTrans = mergedData['transmission'].value_counts().to_dict()
mergedData['transmission_count'] = mergedData['transmission'].map(frequencyDictTrans)
display(mergedData)
```

```python
mergedData = pd.get_dummies(mergedData, columns = ["transmission"])
mergedData = mergedData.drop(['transmission_Amt', 'transmission_Dual-clutch', 'transmission_E-cvt', 'transmission_Ivt'], axis = 1)
mergedData['transmission_Manual'] = mergedData['transmission_Manual'].astype(int)
mergedData['transmission_Automatic'] = mergedData['transmission_Automatic'].astype(int)
mergedData['transmission_CVT'] = mergedData['transmission_CVT'].astype(int)
display(mergedData)
```
