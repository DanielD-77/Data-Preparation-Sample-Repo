# Data Transformation: Primary Dataset



```python
# Show datatypes
display(primaryData.dtypes)
```

```python
# Change the name "price" into "resale_price"
primaryData = primaryData.rename(columns={"price": "resale_price"})
display(primaryData)
```

```python
# Price transformation
def convertPrice(price):
  # If the first letter of price is not
  priceSplit = price.split()
  if price[0] != "₱":
    price = float(price)

  elif len(price.split()) < 2:
    price = price.replace("₱", "")
    price = price.replace(",", "")
    price = float(price)

  else:
    price = priceSplit[0].replace("₱", "")
    price = float(price) * 1000000

  return price

primaryData["resale_price"]= primaryData["resale_price"].apply(convertPrice)
display(primaryData)
```

```python
# Show datatypes
display(primaryData.dtypes)
```

```python
# Transformation from kms_driven data type to float
def convertKms(kms_driven):
  # Checker if kms_driven is not null

  if type(kms_driven) != float:

    kmsSplit = kms_driven.split()
    kms = kmsSplit[0].replace(",", "")
    kms = float(kms)
    return kms

primaryData["kms_driven"] = primaryData["kms_driven"].apply(convertKms)
display(primaryData)
```

```python
# Show datatypes
display(primaryData.dtypes)
```

```python
# Rounded engine size by two decimal places
primaryData["engine_size"] = primaryData["engine_size"].round(2)
display(primaryData)
```
