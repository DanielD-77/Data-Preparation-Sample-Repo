```python
primaryData = pd.read_csv("./datasets/primaryData.csv")
display("Primary Data: %d" % (len(primaryData)))
display(primaryData.head(5))

externalData = pd.read_csv("./datasets/externalData.csv")
display("externalData: %d" % (len(externalData)))
display(externalData.head(5))
```

# Data Cleaning of primaryData

```python
# Check nulls of primaryDataset count
primaryData.isnull().sum()
```

```python
# Drop nulls in kms_driven
primaryData = primaryData.dropna(subset=['kms_driven'])
display(primaryData.isnull().sum())
display(len(primaryData))
```

```python
# We can't do anything with the transmissions here since when buying a car, you get to choose if you want the manual type or automatic type or etc
# So we drop nulls in transmission
primaryData = primaryData.dropna(subset=['transmission'])
display(primaryData.isnull().sum())
display(len(primaryData))
```

```python
# Show nulls in engine_size.
# Upon a quick search in Google, there are cars where you get to choose what
# engine size you want before you buy it, but since there was no engine_size
# listed in the website, we can't look it up and decide the engine_size of those nulls

primaryData[primaryData['engine_size'].isnull()]
```

```python
# Drop nulls in engine_size
primaryData = primaryData.dropna(subset=['engine_size'])
display(primaryData.isnull().sum())
display(len(primaryData))
```

# Data cleaning of externalData

```python
# Show nulls in externalData
externalData.isnull().sum()
```

```python
# Drop the rows that are null in srp of externalDataset
externalData = externalData.dropna(subset=['srp'])
```

```python

display(primaryData.isnull().sum())
display(externalData.isnull().sum())
```
