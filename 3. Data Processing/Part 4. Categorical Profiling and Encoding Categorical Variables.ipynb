```python
# Data Cleaning

df = pd.read_csv("mergedDataset_updated.csv")

def cvt(x):
  if x == "Cvt":
    return "CVT"
  else:
    return x

def body(x):
  if x == "Minivans" or x == "Mpv" or x == "Minivan":
    return "MPV"
  elif x == "Pickup-trucks" or x == "Pickup":
    return "Pickup Truck"
  elif x == "Suv":
    return "SUV"
  elif x == "Luxury-vehicles":
    return "Luxury Vehicle"
  elif x == "Station-wagon":
    return "Station Wagon"
  elif x == "Convertibles":
    return "Convertible"
  else:
    return x


un = df['body_type'].unique()

df['transmission'] = df['transmission'].apply(cvt)

df['body_type'] = df['body_type'].apply(body)

newun = df['body_type'].unique()

newun

display(df)
df.to_csv('mergedDataset_updated.csv', index=False)
```

```python
df = pd.read_csv("mergedDataset_updated.csv")
display(df)
```

```python
# Display the unique fuel_types
df['fuel_type'].unique()
```

```python
# Target Encoding for Categorical Variables

from category_encoders import TargetEncoder

encoder = TargetEncoder()

df['make_target'] = encoder.fit_transform(df['make'], df['resale_price'])
df['model_target'] = encoder.fit_transform(df['model'], df['resale_price'])
df['transmission_target'] = encoder.fit_transform(df['transmission'], df['resale_price'])
df['body_type_target'] = encoder.fit_transform(df['body_type'], df['resale_price'])

df = pd.get_dummies(df, columns=['fuel_type'], dtype=int)
df = df.drop(['fuel_type_Hybrid'], axis=1)

df = df[[c for c in df if c not in ['kms_driven', 'engine_size', 'resale_price', 'srp', 'no_sc_mmnl']] + ['kms_driven', 'engine_size', 'resale_price', 'srp', 'no_sc_mmnl']]

display(df)

```

```python
# Frequency Encoding for Categorical Variables

make_frequency = df['make'].value_counts(normalize=True)

model_frequency = df['model'].value_counts(normalize=True)

transmission_frequency = df['transmission'].value_counts(normalize=True)

body_type_frequency = df['body_type'].value_counts(normalize=True)


df['make_freq'] = df['make'].map(make_frequency)
df['model_freq'] = df['model'].map(model_frequency)
df['transmission_freq'] = df['transmission'].map(transmission_frequency)
df['body_type_freq'] = df['body_type'].map(body_type_frequency)

df = df.drop(['make', 'model', 'transmission', 'body_type'], axis=1)

df = df[[c for c in df if c not in ['fuel_type_Diesel', 'fuel_type_Gasoline', 'kms_driven', 'engine_size', 'resale_price', 'srp', 'no_sc_mmnl']] + ['fuel_type_Diesel', 'fuel_type_Gasoline', 'kms_driven', 'engine_size', 'resale_price', 'srp', 'no_sc_mmnl']]

display(df)

# data.to_csv('FrequencyEncodedDataset.csv', index=False)
```

```python
# Drop key
df = df.drop('key', axis=1)
display(df)
```

```python
# Rearranging the dataset

df = df[["make_year", "ext_make_year", "make_target", "make_freq", "model_target", "model_freq", "transmission_target",
         "transmission_freq", "body_type_target", "body_type_freq", "fuel_type_Diesel",
         "fuel_type_Gasoline", "kms_driven", "engine_size", "resale_price", "srp",
         "no_sc_mmnl"]]

display(df)
```
