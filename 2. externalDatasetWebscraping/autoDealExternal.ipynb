```python
primary = pd.read_csv("./datasets/autodeal.csv")
display(primary)
```

```python
uniqueMakes = primary['Make'].unique()
print(uniqueMakes)
i = 1
for make in uniqueMakes:
  print(f"{i}. {make}")
  i += 1
```

```python
# Webscraping of unique
import requests
import time
from bs4 import BeautifulSoup
import pandas as pd
from pprint import pprint
import numpy as np
```

```python
# Webscraping
card_listings = []
for i in range(11, len(uniqueMakes)):
  print(uniqueMakes[i])

  parent_url = f"https://www.autodeal.com.ph/cars/{uniqueMakes[i]}"

  print(parent_url)

  parent_response = requests.get(parent_url)

  if parent_response.status_code == 200:
      parent_page_content = parent_response.text
  else:
      print("Failed to fetch the parent page. Status code:", parent_response.status_code)

  # Parse the parent page
  parent_soup = BeautifulSoup(parent_page_content, "html.parser")

  # Find all car cards
  car_cards = parent_soup.find_all("div", class_="card-car-model")
  display(len(car_cards))


  for car_card in car_cards:
    #keys and values
    car_card_url = "https://www.autodeal.com.ph%s" % (car_card.find("a")["href"])
    print(car_card_url)
    response = requests.get(car_card_url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        page_content = response.text
    else:
        print("Failed to fetch the page. Status code:", response.status_code)

    soup = BeautifulSoup(page_content, "html.parser")

    # Columns for the Dataframe
    make_model_year = soup.find("h1", class_="bold nomargin padbottom5").text.split()
    make = make_model_year[0]
    model = make_model_year[1]
    year = make_model_year[-1]


    price = soup.find("span", class_="h2").text.split()
    print(price)
    if len(price) == 0:
      srp = np.nan
    else:
      srp = price[1].replace(",", "")
    print(srp)


    # display(make_model_year)

    keys = {"Make": np.nan, "Model": np.nan, "Year": np.nan, "SRP": np.nan}

    keys["Make"] = make
    keys["Model"] = model
    keys["Year"] = year
    keys["SRP"] = srp

    card_listings.append(keys)
    time.sleep(1)

print("Done!")
df = pd.DataFrame(card_listings)
display(df)
```

```python
suzuki_to_kia = pd.DataFrame(card_listings)
display(df)
```

```python
isuzu_to_foton = pd.DataFrame(card_listings)
display(df)
```

```python
# Concatenate both dataframes
combined_df = pd.concat([suzuki_to_kia, isuzu_to_foton])
display(combined_df)
```
