```python
import pandas as pd
import numpy as np
```

```python
primaryData1 = pd.read_csv("./datasets/autodeal.csv")
display("primaryData1: %d" % (len(primaryData1)))
display(primaryData1.head(5))

primaryData2 = pd.read_csv("./datasets/zigwheels.csv")
display("primaryData2: %d" % (len(primaryData2)))
display(primaryData2.head(5))

# Brand new cars external dataset
externalData = pd.read_csv("./datasets/autodealExternal.csv")
display("externalData: %d" % (len(externalData)))
display(externalData)
```

```python
# Drop the unnecessary columns
primaryData2 = primaryData2.drop(["Unnamed: 0", "registration_year", "ownership", "city", "seats"], axis = 1)

display(primaryData2.head(5))
```

```python
# Split model into make, model

def getMake(car):
  deets = car.split(" ")
  make = deets[1]
  return make

def getModel(car):
  deets = car.split(" ")
  model = deets[2]
  return model


primaryData2["Make"] = primaryData2["model"].apply(getMake)
primaryData2["Model"] = primaryData2["model"].apply(getModel)
display(primaryData2.head(5))
```

```python
# Make another column called engine
def getEngine1(car):
  deets = car.split()
  for deet in deets:
    if len(deet) == 4:
      if deet[-1] == "L":
        return deet

  for i in range(len(deets)):
    if len(deets[i]) == 1:
      if deets[i] == "L":
        try:
          float(deets[i-1])
          return deets[i-1]+"L"
        except ValueError:
          continue

  return np.nan
# needs to be able to deal with spaced out version, such as "4.0 L" in the last entry

# for i in range(len(primaryData2)):
#   if primaryData2["engine_displacement"][i].isnull() == float:
#     primaryData2["engine"] = primaryData2["model"].apply(getEngine)
# display(primaryData2)

car_engine = primaryData2["model"].apply(getEngine1)
# display(car_engine)

def getEngine2(engineDis):
  if type(engineDis) != float:
    val = engineDis.split()
    return float(val[0])
  return np.nan

def getEngineFull(car,engineDis):
  if type(getEngine1(car)) == float:
    return getEngine2(engineDis)/1000
  else:
    return float(getEngine1(car)[:-1])

dis_engine = primaryData2["engine_displacement"].apply(getEngine2)
# display(dis_engine)

engine_full = primaryData2.apply(lambda row: getEngineFull(row["model"],row["engine_displacement"]),axis = 1)
# display(engine_full)

# num_null = 0
# for engine in engine_full:
#   if engine*0 != 0:
#     num_null +=1

primaryData2["engine_size"] = engine_full
display(primaryData2.head(5))
```

```python
# Find rows that are null in engine_size
display(primaryData2[primaryData2["engine_size"].isnull()])
```

```python
# Drop the rows that are null
primaryData2 = primaryData2.dropna(subset = ["engine_size"])
display(primaryData2)
```

```python
# Rename column names of primaryData1
primaryData1.rename(columns = {"Make" : "make", "Model" : "model",
                               "Price" : "price", "Vehicle" : "make_year",
                               "Fuel" : "fuel_type", "Mileage" : "kms_driven",
                               "Transmission" : "transmission",
                               "Engine" : "engine_size", "Body" : "body_type"}, inplace = True)
display(primaryData1.head(2))

# Rename column names of primaryData2
primaryData2.rename(columns = {"model" : "car", "Make" : "make", "Model" : "model"}, inplace = True)
display(primaryData2.head(2))
```

```python
# Drop car
primaryData2 = primaryData2.drop(["car"], axis = 1)

# Drop engine_displacement
primaryData2 = primaryData2.drop(["engine_displacement"], axis = 1)
display(primaryData2)
```

```python
# Rearrange primaryData1

primaryData1 = primaryData1[["make", "model", "make_year", "fuel_type",
                             "transmission", "kms_driven", "engine_size",
                             "body_type", "price"]]

display(primaryData1.head(2))

# Rearrange primaryData2

primaryData2 = primaryData2[["make", "model", "make_year", "fuel_type",
                             "transmission", "kms_driven", "engine_size",
                             "body_type", "price"]]
display(primaryData2.head(2))
```

```python
# Rename and rearrange externaldata
externalData.rename(columns = {"Make" : "make", "Model" : "model", "Year" : "make_year", "SRP" : "srp"}, inplace = True)

externalData = externalData[["make", "model", "make_year", "srp"]]
display(externalData.head(2))
```

```python
# Get length of primaryData1 and primaryData2
p1len = len(primaryData1)
p2len = len(primaryData2)

display(f"{p1len} + {p2len} = {p1len + p2len}")
```

```python
# Concatenate primaryData1 and primaryData2
primaryData = pd.concat([primaryData1, primaryData2])
display(primaryData)
```

```python
# Combine Make Model and Year in one column

primaryData["key"] = primaryData["make"] + " " + primaryData["model"]

externalData["key"] = externalData["make"] + " " + externalData["model"]

display(primaryData.head(10))
display(externalData.head(10))
```

```python
# Remove duplicates
# We remove duplicates in the externalDataset
display(len(externalData))

externalData = externalData.drop_duplicates(['key'], keep = 'first')

display(len(externalData))

display(externalData.loc[externalData.duplicated()])
```
