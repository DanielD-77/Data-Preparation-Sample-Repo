```python
import requests
import time
from bs4 import BeautifulSoup
import pandas as pd
from pprint import pprint
import numpy as np
```

```python
# Filtered by cars in price range: 0 - 8,250,000
pages_list = []
for i in range(1, 17):
  parent_url = f"https://www.autodeal.com.ph/used-cars/search/certified-pre-owned+repossessed+used-car-status/8250000-max-price/page-{i}?sort-by=relevance"
  parent_response = requests.get(parent_url)

  # Don't touch anything below this ---------------------------------------------
  if parent_response.status_code == 200:
      parent_page_content = parent_response.text
  else:
      print("Failed to fetch the parent page. Status code:", parent_response.status_code)

  # Parse the parent page
  parent_soup = BeautifulSoup(parent_page_content, "html.parser")

  # Find all car cards
  car_cards = parent_soup.find_all("article", class_="card withshadow row gutters listing-featured padleft padright padbottom")
  pages_list.append(car_cards)
  display(parent_url)
display(len(car_cards))
```

```python
# Initialization of car_cards_urls
card_listings = []
for page in pages_list:
  for car_card in page:

    car_card_url = "https://www.autodeal.com.ph%s" % (car_card.find("a")["href"])
    print(car_card_url)
    response = requests.get(car_card_url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        page_content = response.text
    else:
        print("Failed to fetch the page. Status code:", response.status_code)

    soup = BeautifulSoup(page_content, "html.parser")

    # Columns for the DataFrame
    pad_20_row = soup.find_all("li", class_="pad20 row")
    row_pad20 = soup.find_all("li", class_="row pad20")
    vehicle_age = soup.find("h1", class_="h3 bold nomargin").text.split()[0]


    # Keys and Values
    keys = {"Make": np.nan, "Model": np.nan, "Price": np.nan, "Vehicle" : np.nan,  "Fuel": np.nan, "Transmission": np.nan,  "Mileage": np.nan, "Engine": np.nan, "Body": np.nan}

    keys["Vehicle"] = vehicle_age

    # Extract keys and values from pad_20_row and row_pad20
    for item in row_pad20:
      line = item.text.split()

      if line[0] in keys:
        if line[0] == "Engine" or line[0] == "Fuel":
          keys[line[0]] = line[2]

        elif line[0] == "Mileage":
          keys[line[0]] = line[1]

    # 3
    for item in pad_20_row:
      line = item.text.split()
      if line[0] in keys:
        if line[0] == "Price":
          keys[line[0]] = line[1].replace("P", "").replace(",", "")

        elif line[0] == "Make" or line[0] == "Model":
          keys[line[0]] = line[1]

        elif line[0] == "Body":
          keys[line[0]] = line[2]
        elif line[0] == "Transmission":
          keys[line[0]] = line[1]

    card_listings.append(keys)


    # Append the dictionary to the list
    time.sleep(1)

# -----------------------------------------------------------------------------
# Create a DataFrame from the dictionary
print("Done!")
df = pd.DataFrame(card_listings)
display(df)

```

```python
# All pages from AutoDeal
autodeal_df = df
display(autodeal_df)
```
